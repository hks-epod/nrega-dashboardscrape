```python
#note: due to a bug in python, i had to manually edit the files below to add column headers and convert to "UTF-8 without BOM"
#years = ['2012-2013', '2013-2014', '2014-2015'] # something is wrong with 2012-2013 requests
years = ['2013-2014', '2014-2015']
blocks = {"2012-2013" : pd.read_csv("Blocks1213.csv", index_col=False),
          "2013-2014" : pd.read_csv("Blocks1314.csv", index_col=False),
          "2014-2015" : pd.read_csv("Blocks1415.csv", index_col=False)}

states = {"2012-2013" : pd.read_csv("States1213.csv", index_col=False),
          "2013-2014" : pd.read_csv("States1314.csv", index_col=False),
          "2014-2015" : pd.read_csv("States1415.csv", index_col=False)}

url_stub = 'http://164.100.129.6/netnrega/nrega-reportdashboard/api/dashboard_report_monthly.aspx?'

errors = [] #create list of troublesome state/district/block codes.
rows = []
    
for year in years:
    for state in states[year]["state_code"]: #note that it's not entirely necessary to loop through states
                                             #but doing so allows us to skip states if they are being problematic.
        fail_count = 0
        for block in blocks[year][blocks[year]["state"] == state]["block"]:
            #add preceeding zeroes
            state_code =    str(blocks[year][blocks[year]["block"] == block]["state"].values[0]).zfill(2)
            district_code = str(blocks[year][blocks[year]["block"] == block]["district"].values[0]).zfill(4)
            block_code = str(blocks[year][blocks[year]["block"] == block]["block"].values[0]).zfill(7)

            #create URL
            url = url_stub + 'state_code=' \
                + state_code + '&district_code=' \
                + district_code + '&block_code=' \
                + block_code + '&fin_year=' \
                + year + '&type=b'

            #parse page returns
            try:
                if fail_count <=3:
                    site_text = requests.get(url, timeout=(5.0, 10.0)).text #query and get the text on the site
                    data = ast.literal_eval(site_text)[0] #evalute the text as python code
                    data['state_code'] = state_code
                    data['district_code'] = district_code
                    data['year'] = year
                    rows.append(data)
                    sleep(0.5)
                else:
                    print "State %s is being skipped for year %s for too many failures" % (state_code, year)
                    errors.append([state_code, "ALL", "ALL", year, "ALL"])
                    break

            except requests.exceptions.ConnectTimeout as e:
                print "Server too slow to connect for state %s, district %s, and block %s for year %s" % (state_code, district_code, block_code, year)
                errors.append([state_code, district_code, block_code, year, url])
                fail_count += 1
            except requests.exceptions.ReadTimeout as e:
                print "Server too slow to read for state %s, district %s, and block %s for year %s" % (state_code, district_code, block_code, year)
                errors.append([state_code, district_code, block_code, year, url])
                fail_count += 1
            except:
                print "Could not access page for state %s, district %s, and block %s for year %s" % (state_code, district_code, block_code, year)
                errors.append([state_code, district_code, block_code, year, url])
                fail_count += 1
```

```python
def small_mult_hist(df, var, group, year, title, rows=4, cols=3, bins=10, log=False, thou =False):
    """
    function to make small multiple histograms based on fiscal years

    input
    -----
    
    df: dataframe of interest (NOTE, MUST SORT)
    var: variable to get the distribution of
    group: groups to draw histograms of
    year: fiscal year (ex. '2014-2015')
    title: Super-title of small multiple chart
    rows, cols: how many rows/cols of charts
    bins: how many bins used for histogram
    
    output
    -----
    rows * cols number of graphs.
    
    """
    num_plots = rows * cols  # number of subplots

    figwidth = 20
    figheight = 20

    fig = plt.figure(figsize=(figwidth, figheight))
    fig.suptitle(title, fontsize=30)

    axes = [plt.subplot(rows,cols,i) for i in range(1,num_plots+1)]

    categories = df[(df.year == year)][group].unique()

    for i,x in enumerate(categories):
        ax = axes[i]
        chart_data = df[(df.year == year) & (df[group] == x) & (df[var] != 0)][var].dropna().values
       
        if len(chart_data) >0:
            ax.hist(chart_data, bins=bins, log=log, range=(df[df.year == year][var].min(), df[df.year == year][var].max()) )
            ax.axvline(x=np.median(chart_data),c='red')
            ax.set_title(x)
            if thou: ax.get_xaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int((x*1e-3)), ",") if x>1000 else format(int(x), ",")))
            remove_border(ax)
        #else:
        #    ax.plot([])
            
        
       
    plt.show()
```
